```python
# –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏: 0=admin, 1=guest, 2=editor
# –†–µ—Å—É—Ä—Å—ã: 0=admin_folder, 1=guest_folder, 2=editor_folder
# –ü—Ä–∞–≤–∞: R - —á—Ç–µ–Ω–∏–µ, W - –∑–∞–ø–∏—Å—å, X - –∏—Å–ø–æ–ª–Ω–µ–Ω–∏–µ
acl_matrix = [
    ['RWX', 'R',   'R'  ],  # –ü—Ä–∞–≤–∞ admin'–∞
    ['R',   'R',   ''   ],  # –ü—Ä–∞–≤–∞ guest'–∞ (–æ—à–∏–±–∫–∞: –Ω–µ—Ç –ø—Ä–∞–≤–∞ W –Ω–∞ —Å–≤–æ—é –ø–∞–ø–∫—É)
    ['R',   'RW',  'RWX']   # –ü—Ä–∞–≤–∞ editor'–∞
]

users = ['admin', 'guest', 'editor']
```

**–ó–∞–¥–∞—á–∞:** –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å PowerShell-—Å–∫—Ä–∏–ø—Ç –∏ –Ω–∞–π—Ç–∏ –≤—Å–µ ¬´—Å–ª–æ–≤–∞¬ª (—Å—Ç—Ä–æ–∫–∏ –±–µ–∑ –ø—Ä–æ–±–µ–ª–æ–≤), –∫–æ—Ç–æ—Ä—ã–µ –¥–ª–∏–Ω–Ω–µ–µ 50 —Å–∏–º–≤–æ–ª–æ–≤, —Ç–∞–∫ –∫–∞–∫ —ç—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–º –æ–±—Ñ—É—Å–∫–∞—Ü–∏–∏.

```python
suspicious_script_content = """
function Invoke-Command {
    param($Command)
    $Payload = "aW1wb3J0IG9zCnVzZXJuYW1lID0gb3MuZ2V0bG9naW4oKQpzZXJ2ZXIgPSAnMTkyLjE2OC4xLjEwMCcKcGFzc3dvcmQgPSAnYWRtaW4nCnRyeToKICAgIHNzaCA9IHBhcmFtaWtvL..."
    Invoke-Expression $Command
}
$DecodedPayload = [System.Text.Encoding]::UTF8.GetString([System.Convert]::FromBase64String($Payload))
iex $DecodedPayload
# Final command
Start-Process -FilePath "powershell.exe"
"""
min_length = 50
```

```python
potential_payloads = []

# –†–∞–∑–±–∏–≤–∞–µ–º —Å–∫—Ä–∏–ø—Ç –Ω–∞ "—Å–ª–æ–≤–∞" –ø–æ –ø—Ä–æ–±–µ–ª–∞–º –∏ –ø–µ—Ä–µ–Ω–æ—Å–∞–º —Å—Ç—Ä–æ–∫
words = suspicious_script_content.split()

for word in words:
    # –£–±–∏—Ä–∞–µ–º –∫–∞–≤—ã—á–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–≥—É—Ç –±—ã—Ç—å –≤ –Ω–∞—á–∞–ª–µ –∏–ª–∏ –∫–æ–Ω—Ü–µ —Å–ª–æ–≤–∞
    cleaned_word = word.strip('""')

    if len(cleaned_word) > min_length:
        print(f"[!] –û–±–Ω–∞—Ä—É–∂–µ–Ω–∞ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–æ –≤—Ä–µ–¥–æ–Ω–æ—Å–Ω–∞—è —Å—Ç—Ä–æ–∫–∞ –¥–ª–∏–Ω–æ–π {len(cleaned_word)} —Å–∏–º–≤–æ–ª–æ–≤.")
        potential_payloads.append(cleaned_word)

print("\n--- –†–µ–∑—É–ª—å—Ç–∞—Ç –∞–Ω–∞–ª–∏–∑–∞ ---")
if potential_payloads:
    print(f"–ù–∞–π–¥–µ–Ω–Ω—ã–µ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–µ payload'—ã: {potential_payloads}")
else:
    print("–ê–Ω–æ–º–∞–ª—å–Ω–æ –¥–ª–∏–Ω–Ω—ã—Ö —Å—Ç—Ä–æ–∫ –Ω–µ –Ω–∞–π–¥–µ–Ω–æ.")
```

–¢–µ–ø–µ—Ä—å –æ–±—ä–µ–¥–∏–Ω–∏–º –≤—Å—ë, —á—Ç–æ –º—ã –∏–∑—É—á–∏–ª–∏: —Ü–∏–∫–ª—ã, —É—Å–ª–æ–≤–∏—è, —Ä–∞–±–æ—Ç—É —Å–æ —Å—Ç—Ä–æ–∫–∞–º–∏ –∏ —Å–ø–∏—Å–∫–∏.

–ù–∞–ø–∏—à–µ–º –ø—Ä–æ—Å—Ç—É—é, –Ω–æ –ø–æ–ª–µ–∑–Ω—É—é —É—Ç–∏–ª–∏—Ç—É ‚Äî –≤–∞–ª–∏–¥–∞—Ç–æ—Ä –ø–∞—Ä–æ–ª–µ–π, –æ–Ω –±—É–¥–µ—Ç –ø—Ä–æ–≤–µ—Ä—è—Ç—å –ø–∞—Ä–æ–ª—å –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –±–∞–∑–æ–≤—ã–º –ø–æ–ª–∏—Ç–∏–∫–∞–º –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç–∏.

```python
password_to_check = input("–í–≤–µ–¥–∏—Ç–µ –ø–∞—Ä–æ–ª—å –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏: ")
errors = []

# 1. –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–ª–∏–Ω—ã
if len(password_to_check) < 8:
    errors.append("–ü–∞—Ä–æ–ª—å —Å–ª–∏—à–∫–æ–º –∫–æ—Ä–æ—Ç–∫–∏–π (–º–µ–Ω–µ–µ 8 —Å–∏–º–≤–æ–ª–æ–≤).")

# 2. –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –Ω–∞–ª–∏—á–∏–µ —Ü–∏—Ñ—Ä
has_digit = False
for char in password_to_check:
    if char.isdigit():
        has_digit = True
        break # –ù–∞—à–ª–∏ —Ü–∏—Ñ—Ä—É, –¥–∞–ª—å—à–µ –∏—Å–∫–∞—Ç—å –Ω–µ—Ç —Å–º—ã—Å–ª–∞
if not has_digit:
    errors.append("–ü–∞—Ä–æ–ª—å –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ö–æ—Ç—è –±—ã –æ–¥–Ω—É —Ü–∏—Ñ—Ä—É.")

# 3. –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –∑–∞–≥–ª–∞–≤–Ω—ã—Ö –±—É–∫–≤
has_upper = False
for char in password_to_check:
    if char.isupper():
        has_upper = True
        break
if not has_upper:
    errors.append("–ü–∞—Ä–æ–ª—å –¥–æ–ª–∂–µ–Ω —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ö–æ—Ç—è –±—ã –æ–¥–Ω—É –∑–∞–≥–ª–∞–≤–Ω—É—é –±—É–∫–≤—É.")

# 4. –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print("\n--- –†–µ–∑—É–ª—å—Ç–∞—Ç –ø—Ä–æ–≤–µ—Ä–∫–∏ ---")
if not errors:
    print("[OK] –ü–∞—Ä–æ–ª—å —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –±–∞–∑–æ–≤—ã–º —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è–º –±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç–∏.")
else:
    print("[!] –ü–∞—Ä–æ–ª—å –Ω–µ–Ω–∞–¥—ë–∂–µ–Ω. –û–±–Ω–∞—Ä—É–∂–µ–Ω—ã —Å–ª–µ–¥—É—é—â–∏–µ –ø—Ä–æ–±–ª–µ–º—ã:")
    # –í—ã–≤–æ–¥–∏–º –≤—Å–µ –Ω–∞–π–¥–µ–Ω–Ω—ã–µ –æ—à–∏–±–∫–∏
    for error in errors:
        print(f"- {error}")
```

## –í–∞—à–∏ –≤–æ–ø—Ä–æ—Å—ã

### 2. –¶–∏–∫–ª `while`: –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–µ –ø–æ —É—Å–ª–æ–≤–∏—é

–í –æ—Ç–ª–∏—á–∏–µ –æ—Ç —Ü–∏–∫–ª–∞ `for`, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–æ—Ö–æ–¥–∏—Ç –ø–æ —ç–ª–µ–º–µ–Ω—Ç–∞–º –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏, —Ü–∏–∫–ª `while` (—Å –∞–Ω–≥–ª. –ø–æ–∫–∞) –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ **—É—Å–ª–æ–≤–∏–µ –∏—Å—Ç–∏–Ω–Ω–æ (`True`)**. –ö–∞–∫ —Ç–æ–ª—å–∫–æ —É—Å–ª–æ–≤–∏–µ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –ª–æ–∂–Ω—ã–º (`False`), —Ü–∏–∫–ª –ø—Ä–µ–∫—Ä–∞—â–∞–µ—Ç—Å—è.

**–ö–ª—é—á–µ–≤–æ–µ –æ—Ç–ª–∏—á–∏–µ –æ—Ç `for`:** —Ü–∏–∫–ª `while` –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ç–æ–≥–¥–∞, –∫–æ–≥–¥–∞ –Ω–µ –∏–∑–≤–µ—Å—Ç–Ω–æ –∑–∞—Ä–∞–Ω–µ–µ, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –Ω—É–∂–Ω–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å –∏—Ç–µ—Ä–∞—Ü–∏—é, –Ω–æ –∏–∑–≤–µ—Å—Ç–Ω–æ —É—Å–ª–æ–≤–∏–µ, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –Ω—É–∂–Ω–æ –ø—Ä–æ–¥–æ–ª–∂–∞—Ç—å –∏–ª–∏ –æ—Å—Ç–∞–Ω–æ–≤–∏—Ç—å—Å—è.




–°–∏–º—É–ª—è—Ü–∏—è —Å–∏—Å—Ç–µ–º—ã –∑–∞—â–∏—Ç—ã –æ—Ç brute-force –∞—Ç–∞–∫–∏:

```python
# –£—Å–ª–æ–≤–∏—è –≤—Ö–æ–¥–∞
correct_password = "SecretPassword123!"
attempts_left = 3
is_locked = False
login_successful = False

print("--- –°–∏—Å—Ç–µ–º–∞ –≤—Ö–æ–¥–∞ –≤ –∑–∞—â–∏—â–µ–Ω–Ω—ã–π —Å–µ—Ä–≤–µ—Ä ---")

while attempts_left > 0:
    print(f"–£ –≤–∞—Å –æ—Å—Ç–∞–ª–æ—Å—å –ø–æ–ø—ã—Ç–æ–∫: {attempts_left}")
    entered_password = input("–í–≤–µ–¥–∏—Ç–µ –ø–∞—Ä–æ–ª—å: ")

    if entered_password == correct_password:
        print("\n[OK] –î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω!")
        login_successful = True # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Ñ–ª–∞–≥ —É—Å–ø–µ—Ö–∞
        break # –í—ã—Ö–æ–¥–∏–º –∏–∑ —Ü–∏–∫–ª–∞, –µ—Å–ª–∏ –ø–∞—Ä–æ–ª—å –≤–µ—Ä–Ω—ã–π
    else:
        print("[X] –ù–µ–≤–µ—Ä–Ω—ã–π –ø–∞—Ä–æ–ª—å.")
        attempts_left -= 1

# –ü–æ—Å–ª–µ —Ü–∏–∫–ª–∞ –ø—Ä–æ–≤–µ—Ä—è–µ–º, –ø–æ—á–µ–º—É –º—ã –∏–∑ –Ω–µ–≥–æ –≤—ã—à–ª–∏.
# –ï—Å–ª–∏ –≤—ã—à–ª–∏ –ø–æ break, —Ç–æ login_successful –±—É–¥–µ—Ç True.
# –ï—Å–ª–∏ –∫–æ–Ω—á–∏–ª–∏—Å—å –ø–æ–ø—ã—Ç–∫–∏, —Ç–æ login_successful –æ—Å—Ç–∞–Ω–µ—Ç—Å—è False.
if not login_successful:
    is_locked = True
    print("\n[!!!] –ê–ö–ö–ê–£–ù–¢ –ó–ê–ë–õ–û–ö–ò–†–û–í–ê–ù! –°–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ –Ω–µ—É–¥–∞—á–Ω—ã—Ö –ø–æ–ø—ã—Ç–æ–∫.")
```

**–°–∏–º—É–ª—è—Ü–∏—è –∑–∞–≤–∏—Å—à–µ–≥–æ –ø—Ä–æ—Ü–µ—Å—Å–∞ (–±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–π —Ü–∏–∫–ª)**

**–í–∞–∂–Ω–æ:** –∏–Ω–æ–≥–¥–∞ –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–µ —Ü–∏–∫–ª—ã –≤–æ–∑–Ω–∏–∫–∞—é—Ç —Å–ª—É—á–∞–π–Ω–æ –∏ –ø—Ä–∏–≤–æ–¥—è—Ç –∫ –æ—Ç–∫–∞–∑—É –≤ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏–∏ (DoS). –í–∞—à–∞ –∑–∞–¥–∞—á–∞ –∫–∞–∫ —Å–ø–µ—Ü–∏–∞–ª–∏—Å—Ç–∞ ‚Äî —É–º–µ—Ç—å –∏—Ö —Ä–∞—Å–ø–æ–∑–Ω–∞–≤–∞—Ç—å –∏ –ø—Ä–µ–¥–æ—Ç–≤—Ä–∞—â–∞—Ç—å.

**–°—Ü–µ–Ω–∞—Ä–∏–π:** —Å–∫—Ä–∏–ø—Ç –∂–¥—ë—Ç –æ—Ç–≤–µ—Ç–∞ –æ—Ç —É–¥–∞–ª—ë–Ω–Ω–æ–≥–æ —Å–µ—Ä–≤–µ—Ä–∞, –µ—Å–ª–∏ —Å–µ—Ä–≤–µ—Ä –Ω–µ –æ—Ç–≤–µ—á–∞–µ—Ç, —Å–∫—Ä–∏–ø—Ç –º–æ–∂–µ—Ç –∑–∞–≤–∏—Å–Ω—É—Ç—å –Ω–∞–≤—Å–µ–≥–¥–∞.

```python
server_is_online = False # –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –ø–æ–º–µ–Ω—è—Ç—å –Ω–∞ True

print("–ü–æ–ø—ã—Ç–∫–∞ –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è –∫ —Å–µ—Ä–≤–µ—Ä—É...")
while not server_is_online:
    # –í —Ä–µ–∞–ª—å–Ω–æ–º –∫–æ–¥–µ –∑–¥–µ—Å—å –±—ã–ª–∞ –±—ã –ø–æ–ø—ã—Ç–∫–∞ —Å–µ—Ç–µ–≤–æ–≥–æ —Å–æ–µ–¥–∏–Ω–µ–Ω–∏—è.
    # –¢–∞–∫ –∫–∞–∫ —Å—Ç–∞—Ç—É—Å –Ω–µ –º–µ–Ω—è–µ—Ç—Å—è, –µ—Å–ª–∏ —Å–µ—Ä–≤–µ—Ä –æ—Ñ—Ñ–ª–∞–π–Ω, –º—ã –∑–∞—Ü–∏–∫–ª–∏–º—Å—è –Ω–∞–≤—Å–µ–≥–¥–∞!
    print("–°–µ—Ä–≤–µ—Ä –Ω–µ –æ—Ç–≤–µ—á–∞–µ—Ç, –ø—Ä–æ–±—É—é —Å–Ω–æ–≤–∞...")
    # time.sleep(1) # –ó–∞–¥–µ—Ä–∂–∫–∞, —á—Ç–æ–±—ã –Ω–µ –∑–∞–≥—Ä—É–∂–∞—Ç—å –ø—Ä–æ—Ü–µ—Å—Å–æ—Ä –Ω–∞ 100%
print("–°–µ—Ä–≤–µ—Ä –æ–Ω–ª–∞–π–Ω!")
```

```python
max_retries = 5
retries = 0
server_is_online = False

print("\n–ü–æ–ø—ã—Ç–∫–∞ –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è –∫ —Å–µ—Ä–≤–µ—Ä—É (—Å —Ç–∞–π–º–∞—É—Ç–æ–º)...")
while not server_is_online and retries < max_retries:
    print(f"–°–µ—Ä–≤–µ—Ä –Ω–µ –æ—Ç–≤–µ—á–∞–µ—Ç, –ø—Ä–æ–±—É—é —Å–Ω–æ–≤–∞ (–ø–æ–ø—ã—Ç–∫–∞ {retries + 1} –∏–∑ {max_retries})...")
    retries += 1
    # time.sleep(1)

if server_is_online:
    print("–°–µ—Ä–≤–µ—Ä –æ–Ω–ª–∞–π–Ω!")
else:
    print("–ù–µ —É–¥–∞–ª–æ—Å—å –ø–æ–¥–∫–ª—é—á–∏—Ç—å—Å—è –∫ —Å–µ—Ä–≤–µ—Ä—É –ø–æ—Å–ª–µ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –ø–æ–ø—ã—Ç–æ–∫.")
```

–í—Å–µ–≥–¥–∞ —É–±–µ–∂–¥–∞–π—Ç–µ—Å—å, —á—Ç–æ –≤–Ω—É—Ç—Ä–∏ —Ç–µ–ª–∞ —Ü–∏–∫–ª–∞ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç —á—Ç–æ-—Ç–æ, —á—Ç–æ –≤ –∫–æ–Ω–µ—á–Ω–æ–º –∏—Ç–æ–≥–µ —Å–¥–µ–ª–∞–µ—Ç `—É—Å–ª–æ–≤–∏–µ` –ª–æ–∂–Ω—ã–º. –û–±—ã—á–Ω–æ —ç—Ç–æ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ–≤–µ—Ä—è–µ—Ç—Å—è –≤ —É—Å–ª–æ–≤–∏–∏.

## –í–∞—à–∏ –≤–æ–ø—Ä–æ—Å—ã

### 3. –ë–ª–æ–∫ `else` –¥–ª—è —Ü–∏–∫–ª–æ–≤

–ë–ª–æ–∫ `else`, –Ω–∞–ø–∏—Å–∞–Ω–Ω—ã–π –Ω–∞ —Ç–æ–º –∂–µ —É—Ä–æ–≤–Ω–µ –æ—Ç—Å—Ç—É–ø–∞, —á—Ç–æ –∏ `for` –∏–ª–∏ `while`, –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –≤ —Ç–æ–º —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ —Ü–∏–∫–ª –∑–∞–≤–µ—Ä—à–∏–ª—Å—è **–µ—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–º –æ–±—Ä–∞–∑–æ–º** (—Ç.–µ. –±—ã–ª–∏ –ø—Ä–æ–π–¥–µ–Ω—ã –≤—Å–µ –∏—Ç–µ—Ä–∞—Ü–∏–∏ `for`, –∏–ª–∏ —É—Å–ª–æ–≤–∏–µ `while` —Å—Ç–∞–ª–æ –ª–æ–∂–Ω—ã–º), **–∞ –Ω–µ –±—ã–ª –ø—Ä–µ—Ä–≤–∞–Ω** –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–º `break`.


*   **–¶–∏–∫–ª `for`:** –µ—Å–ª–∏ –º—ã –∏—â–µ–º —ç–ª–µ–º–µ–Ω—Ç –∏ –Ω–∞—Ö–æ–¥–∏–º –µ–≥–æ (`break`), –±–ª–æ–∫ `else` –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è; –µ—Å–ª–∏ –≤—ã –ø—Ä–æ—à–ª–∏ –≤–µ—Å—å —Å–ø–∏—Å–æ–∫ –∏ –Ω–µ –Ω–∞—à–ª–∏ (—Ü–∏–∫–ª –∑–∞–≤–µ—Ä—à–∏–ª—Å—è —Å–∞–º), `else` –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è.
*   **–¶–∏–∫–ª `while`:** –µ—Å–ª–∏ —É—Å–ª–æ–≤–∏–µ —Ü–∏–∫–ª–∞ —Å—Ç–∞–ª–æ –ª–æ–∂–Ω—ã–º –∏ –≤—ã –≤—ã—à–ª–∏, `else` –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è, –µ—Å–ª–∏ –≤—ã –≤—ã—à–ª–∏ –∏–∑ —Ü–∏–∫–ª–∞ –ø–æ `break`, `else` –Ω–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è.


```python
# –£—Å–ª–æ–≤–∏—è –≤—Ö–æ–¥–∞
correct_password = "SecretPassword123!"
attempts_left = 3
is_locked = False

print("--- –°–∏—Å—Ç–µ–º–∞ –≤—Ö–æ–¥–∞ –≤ –∑–∞—â–∏—â–µ–Ω–Ω—ã–π —Å–µ—Ä–≤–µ—Ä ---")

while attempts_left > 0:
    print(f"–£ –≤–∞—Å –æ—Å—Ç–∞–ª–æ—Å—å –ø–æ–ø—ã—Ç–æ–∫: {attempts_left}")
    entered_password = input("–í–≤–µ–¥–∏—Ç–µ –ø–∞—Ä–æ–ª—å: ")

    if entered_password == correct_password:
        print("\n[OK] –î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω!")
        break # –í—ã—Ö–æ–¥–∏–º –∏–∑ —Ü–∏–∫–ª–∞, –µ—Å–ª–∏ –ø–∞—Ä–æ–ª—å –≤–µ—Ä–Ω—ã–π
    else:
        print("[X] –ù–µ–≤–µ—Ä–Ω—ã–π –ø–∞—Ä–æ–ª—å.")
        attempts_left -= 1
# –ë–ª–æ–∫ else —Å—Ä–∞–±–æ—Ç–∞–µ—Ç, –µ—Å–ª–∏ —Ü–∏–∫–ª –∑–∞–≤–µ—Ä—à–∏–ª—Å—è —Å–∞–º (–ø–æ–ø—ã—Ç–∫–∏ –∫–æ–Ω—á–∏–ª–∏—Å—å), –∞ –Ω–µ —á–µ—Ä–µ–∑ break
else:
    is_locked = True
    print("\n[!!!] –ê–ö–ö–ê–£–ù–¢ –ó–ê–ë–õ–û–ö–ò–†–û–í–ê–ù! –°–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ –Ω–µ—É–¥–∞—á–Ω—ã—Ö –ø–æ–ø—ã—Ç–æ–∫.")
```

## –í–∞—à–∏ –≤–æ–ø—Ä–æ—Å—ã

### 4. –í–ª–æ–∂–µ–Ω–Ω—ã–µ —Ü–∏–∫–ª—ã

–ò–Ω–æ–≥–¥–∞ –¥–ª—è —Ä–µ—à–µ–Ω–∏—è –∑–∞–¥–∞—á–∏ –Ω—É–∂–Ω–æ –ø–æ–º–µ—Å—Ç–∏—Ç—å –æ–¥–∏–Ω —Ü–∏–∫–ª –≤–Ω—É—Ç—Ä—å –¥—Ä—É–≥–æ–≥–æ. –≠—Ç–æ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è **–≤–ª–æ–∂–µ–Ω–Ω—ã–º —Ü–∏–∫–ª–æ–º (nested loop)**.

**–ö–∞–∫ —ç—Ç–æ —Ä–∞–±–æ—Ç–∞–µ—Ç**

–î–ª—è –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ –≤–Ω–µ—à–Ω–µ–≥–æ —Ü–∏–∫–ª–∞, –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π —Ü–∏–∫–ª –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –ø–æ–ª–Ω–æ—Å—Ç—å—é –æ—Ç –Ω–∞—á–∞–ª–∞ –¥–æ –∫–æ–Ω—Ü–∞.

–í–ª–æ–∂–µ–Ω–Ω—ã–µ —Ü–∏–∫–ª—ã –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –Ω—É–∂–Ω—ã –≤ –¥–≤—É—Ö —Å–∏—Ç—É–∞—Ü–∏—è—Ö:

- **–≥–µ–Ω–µ—Ä–∞—Ü–∏—è –∫–æ–º–±–∏–Ω–∞—Ü–∏–π** ‚Äî –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–æ —Å–∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã –∏–∑ –¥–≤—É—Ö –∏–ª–∏ –±–æ–ª–µ–µ –∫–æ–ª–ª–µ–∫—Ü–∏–π;
- **–æ–±—Ä–∞–±–æ—Ç–∫–∞ –º–Ω–æ–≥–æ–º–µ—Ä–Ω—ã—Ö —Å—Ç—Ä—É–∫—Ç—É—Ä** ‚Äî –∫–æ–≥–¥–∞ –¥–∞–Ω–Ω—ã–µ –æ—Ä–≥–∞–Ω–∏–∑–æ–≤–∞–Ω—ã –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü—ã —Å–ø–∏—Å–∫–∞ —Å–ø–∏—Å–∫–æ–≤ –∏–ª–∏ –¥—Ä—É–≥–æ–π –≤–ª–æ–∂–µ–Ω–Ω–æ–π —Å—Ç—Ä—É–∫—Ç—É—Ä—ã.

–†–∞–∑–±–µ—Ä—ë–º, –∫–∞–∫ –≤–ª–æ–∂–µ–Ω–Ω—ã–µ —Ü–∏–∫–ª—ã –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –¥–ª—è —Å–∏–º—É–ª—è—Ü–∏–∏ **–∞—Ç–∞–∫–∏ –º–µ—Ç–æ–¥–æ–º –ø–µ—Ä–µ–±–æ—Ä–∞ (brute-force)** –ø–æ —Å–ª–æ–≤–∞—Ä—é.

**–ó–∞–¥–∞—á–∞:** —Å–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ –ø–∞—Ä–æ–ª–∏, –∫–æ–º–±–∏–Ω–∏—Ä—É—è –±–∞–∑–æ–≤—ã–µ —Å–ª–æ–≤–∞ –∏ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ —á–∏—Å–ª–æ–≤—ã–µ –æ–∫–æ–Ω—á–∞–Ω–∏—è.

```python
# –°–ø–∏—Å–∫–∏ –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Å–ª–æ–≤–∞—Ä—è –ø–∞—Ä–æ–ª–µ–π
potential_usernames = ['admin', 'root']
common_passwords_parts = ['password', '123456', 'qwerty']

print("--- –ù–∞—á–∏–Ω–∞—é –≥–µ–Ω–µ—Ä–∞—Ü–∏—é —Å–ª–æ–≤–∞—Ä—è –¥–ª—è –∞—Ç–∞–∫–∏ ---")
attack_combo_count = 0

# –í–Ω–µ—à–Ω–∏–π —Ü–∏–∫–ª –ø–æ –∏–º–µ–Ω–∞–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
for user in potential_usernames:
    # –í–Ω—É—Ç—Ä–µ–Ω–Ω–∏–π —Ü–∏–∫–ª –ø–æ —á–∞—Å—Ç—è–º –ø–∞—Ä–æ–ª—è
    for password_part in common_passwords_parts:
        # –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –∫–æ–º–±–∏–Ω–∞—Ü–∏—é
        generated_password = f"{user}{password_part}"
        print(f"–ü–æ–ø—ã—Ç–∫–∞ –≤—Ö–æ–¥–∞: –ü–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å='{user}', –ü–∞—Ä–æ–ª—å='{generated_password}'")
        attack_combo_count += 1

print(f"\n–í—Å–µ–≥–æ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–æ {attack_combo_count} –∫–æ–º–±–∏–Ω–∞—Ü–∏–π –¥–ª—è –∞—Ç–∞–∫–∏.")
```

**–í–∞–∂–Ω–æ**

*   **–ü—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å** ‚Äî –ø–æ–º–Ω–∏—Ç–µ, —á—Ç–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø–µ—Ä–∞—Ü–∏–π –≤–æ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–µ–º —Ü–∏–∫–ª–µ —Ä–∞–≤–Ω–æ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—é —á–∏—Å–ª–∞ –∏—Ç–µ—Ä–∞—Ü–∏–π –≤–Ω–µ—à–Ω–µ–≥–æ –∏ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–µ–≥–æ —Ü–∏–∫–ª–æ–≤ (N*M), –¥–ª—è –±–æ–ª—å—à–∏—Ö N –∏ M —ç—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å –º–µ–¥–ª–µ–Ω–Ω–æ.
*   **–ß–∏—Ç–∞–µ–º–æ—Å—Ç—å** ‚Äî –≥–ª—É–±–æ–∫–∞—è –≤–ª–æ–∂–µ–Ω–Ω–æ—Å—Ç—å (3+ —É—Ä–æ–≤–Ω—è) —Å–∏–ª—å–Ω–æ –∑–∞—Ç—Ä—É–¥–Ω—è–µ—Ç —á—Ç–µ–Ω–∏–µ –∫–æ–¥–∞, —Å—Ç–∞—Ä–∞–π—Ç–µ—Å—å –∏–∑–±–µ–≥–∞—Ç—å —ç—Ç–æ–≥–æ, –≤–æ–∑–º–æ–∂–Ω–æ, —Ä–∞–∑–±–∏–≤–∞—è –ª–æ–≥–∏–∫—É –Ω–∞ —á–∞—Å—Ç–∏, –≤ –±—É–¥—É—â–µ–º ‚Äî –Ω–∞ —Ñ—É–Ω–∫—Ü–∏–∏.
*   **`break`/`continue`** ‚Äî –¥–µ–π—Å—Ç–≤—É—é—Ç —Ç–æ–ª—å–∫–æ –Ω–∞ —Ç–æ—Ç —Ü–∏–∫–ª, –≤–Ω—É—Ç—Ä–∏ –∫–æ—Ç–æ—Ä–æ–≥–æ –æ–Ω–∏ –Ω–∞–ø–∏—Å–∞–Ω—ã, `break` –≤–æ –≤–Ω—É—Ç—Ä–µ–Ω–Ω–µ–º —Ü–∏–∫–ª–µ –Ω–µ –ø—Ä–µ—Ä–≤–µ—Ç –≤–Ω–µ—à–Ω–∏–π.

**–°–∏—Ç—É–∞—Ü–∏—è:** –µ—Å—Ç—å –æ—Ç—á—ë—Ç—ã –æ—Ç —Ç—Ä—ë—Ö —Ä–∞–∑–Ω—ã—Ö –∞–Ω—Ç–∏–≤–∏—Ä—É—Å–Ω—ã—Ö —Å–∫–∞–Ω–µ—Ä–æ–≤, –∫–∞–∂–¥—ã–π –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –≤–µ—Ä–Ω—É–ª —Å–≤–æ–π —Å–ø–∏—Å–æ–∫ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–Ω—ã—Ö –≤—Ä–µ–¥–æ–Ω–æ—Å–Ω—ã—Ö —Ñ–∞–π–ª–æ–≤.

**–ó–∞–¥–∞—á–∞:** –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Å–æ–∑–¥–∞—Ç—å –µ–¥–∏–Ω—ã–π —Å–ø–∏—Å–æ–∫ –≤—Å–µ—Ö –æ–±–Ω–∞—Ä—É–∂–µ–Ω–Ω—ã—Ö —Ñ–∞–π–ª–æ–≤.




```python
# –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ—Ç —Ä–∞–∑–Ω—ã—Ö —Å–∫–∞–Ω–µ—Ä–æ–≤
scanner_results = [
    ['virus.exe', 'trojan.dll'],                  # –û—Ç—á–µ—Ç –æ—Ç Kaspersky
    ['adware.js', 'virus.exe'],                   # –û—Ç—á–µ—Ç –æ—Ç Dr.Web
    ['keylogger.py', 'rootkit.sys', 'trojan.dll'] # –û—Ç—á–µ—Ç –æ—Ç Avast
]

# –¥–æ–ª–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å—Å—è:
# ['virus.exe', 'trojan.dll', 'adware.js', 'virus.exe', 'keylogger.py', 'rootkit.sys','trojan.dll']
```

**–ó–∞–¥–∞—á–∞**

–ï—Å—Ç—å —Ñ—Ä–∞–≥–º–µ–Ω—Ç –ª–æ–≥–∞ –∞—É—Ç–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ü–∏–∏.

–í–∞—à–∞ —Ü–µ–ª—å ‚Äî –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —ç—Ç–æ—Ç —Ñ—Ä–∞–≥–º–µ–Ω—Ç, –Ω–∞–π—Ç–∏ –≤—Å–µ –Ω–µ—É–¥–∞—á–Ω—ã–µ –ø–æ–ø—ã—Ç–∫–∏ –≤—Ö–æ–¥–∞ –∏ —Å–æ—Å—Ç–∞–≤–∏—Ç—å —Å–ø–∏—Å–æ–∫ IP-–∞–¥—Ä–µ—Å–æ–≤, —Å –∫–æ—Ç–æ—Ä—ã—Ö —ç—Ç–∏ –ø–æ–ø—ã—Ç–∫–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–∏–ª–∏—Å—å, —á—Ç–æ–±—ã –ø–µ—Ä–µ–¥–∞—Ç—å –∏—Ö –≤ —Å–∏—Å—Ç–µ–º—É –±–ª–æ–∫–∏—Ä–æ–≤–∫–∏.

–ï—Å–ª–∏ —É –≤–∞—Å –µ—Å—Ç—å –¥–≤–∞ —Å–ø–∏—Å–∫–∞ **–æ–¥–∏–Ω–∞–∫–æ–≤–æ–π –¥–ª–∏–Ω—ã**, –∏ —ç–ª–µ–º–µ–Ω—Ç—ã –≤ –Ω–∏—Ö —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –¥—Ä—É–≥ –¥—Ä—É–≥—É –ø–æ –∏–Ω–¥–µ–∫—Å—É (–ø–µ—Ä–≤—ã–π —Å –ø–µ—Ä–≤—ã–º, –≤—Ç–æ—Ä–æ–π —Å–æ –≤—Ç–æ—Ä—ã–º –∏ —Ç.–¥.), –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤–ª–æ–∂–µ–Ω–Ω—ã–π —Ü–∏–∫–ª ‚Äî –æ—à–∏–±–∫–∞. –≠—Ç–æ –Ω–µ—ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ –∏ —É—Å–ª–æ–∂–Ω—è–µ—Ç –∫–æ–¥.

–î–ª—è —Ç–∞–∫–∏—Ö –∑–∞–¥–∞—á —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —ç–ª–µ–≥–∞–Ω—Ç–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ ‚Äî —Ñ—É–Ω–∫—Ü–∏—è `zip()`.

**–ó–∞–¥–∞—á–∞:** –µ—Å—Ç—å —Å–ø–∏—Å–æ–∫ —Ñ–∞–π–ª–æ–≤ –∏ –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã–π —Å–ø–∏—Å–æ–∫ –∏—Ö MD5-—Ö—ç—à–µ–π, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–≤–µ—Å—Ç–∏ –∏—Ö –ø–∞—Ä–∞–º–∏: `—Ñ–∞–π–ª -> —Ö—ç—à`.

```python
filenames = ['kernel32.dll', 'user32.dll', 'svchost.exe']
hashes = ['d2698934b87572d4b8b6a4ba21e1a2d5', 'f8de9210a4e8d0e7a5e8f4d9c7d0b3e1', 'a3c5d8a9f0e1b2c3d4e5f6a7b8c9d0e1']

# --- –ù–µ—ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–π —Å–ø–æ—Å–æ–± (—á–µ—Ä–µ–∑ –≤–ª–æ–∂–µ–Ω–Ω—ã–µ —Ü–∏–∫–ª—ã –∏ –∏–Ω–¥–µ–∫—Å—ã) ---
# –≠—Ç–æ—Ç –∫–æ–¥ —Ä–∞–±–æ—Ç–∞–µ—Ç, –Ω–æ –æ–Ω –æ—á–µ–Ω—å –≥—Ä–æ–º–æ–∑–¥–∫–∏–π –∏ —Å–ª–æ–∂–Ω—ã–π –¥–ª—è —á—Ç–µ–Ω–∏—è.
print("--- –í—ã–≤–æ–¥ —á–µ—Ä–µ–∑ –≤–ª–æ–∂–µ–Ω–Ω—ã–µ —Ü–∏–∫–ª—ã (–Ω–µ –¥–µ–ª–∞–π—Ç–µ —Ç–∞–∫) ---")
i = 0
for file in filenames:
    j = 0
    for h in hashes:
        if i == j:
            print(f"{file} -> {h}")
        j += 1
    i += 1

```

```python
# --- –≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã–π –ø–∏—Ç–æ–Ω–∏—á–Ω—ã–π —Å–ø–æ—Å–æ–± (—á–µ—Ä–µ–∑ zip) ---
# zip() —Å–∫–ª–µ–∏–≤–∞–µ—Ç —Å–ø–∏—Å–∫–∏, —Å–æ–∑–¥–∞–≤–∞—è –Ω–∞ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏ –∫–æ—Ä—Ç–µ–∂ –∏–∑ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤.
print("\n--- –í—ã–≤–æ–¥ —á–µ—Ä–µ–∑ zip (–ü–†–ê–í–ò–õ–¨–ù–´–ô –°–ü–û–°–û–ë) ---")
for file, file_hash in zip(filenames, hashes):
    print(f"{file} -> {file_hash}")
```

- –ù–∞—É—á–∏–ª–∏—Å—å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ü–∏–∫–ª—ã for –∏ while.
