Реализуйте функцию sum_distance(from, to), которая суммирует все числа от
значения from до величины to включительно.
Примечание. Если пользователь задаст первое число, которое окажется

```python
def sum_distance(from_value, to_value):
    """
    Суммирует все числа от from_value до to_value включительно.

    Если from_value больше to_value, значения будут автоматически поменяны местами.

    Args:
        from_value (int): Начальное число диапазона.
        to_value (int): Конечное число диапазона.

    Returns:
        int: Сумма всех чисел в указанном диапазоне.
    """
    # Шаг 1: Проверяем и меняем местами значения, если from больше to
    if from_value > to_value:
        from_value, to_value = to_value, from_value
        print(f"Примечание: Первое число было больше второго. Диапазон поменян на {from_value} до {to_value}.")

    # Шаг 2: Суммируем все числа в диапазоне
    # range(start, end) генерирует числа от start до end-1.
    # Чтобы включить to, мы должны указать to + 1.
    total_sum = sum(range(from_value, to_value + 1))

    return total_sum
```

Напишите код на Python для решения следующей задачи.
Реализуйте функцию trim_and_repeat(), которая принимает три параметра:
● строку;
● offset — число символов, на которое нужно обрезать строку слева;
● repetitions — сколько раз нужно повторить строку перед возвратом
получившейся строки.
Число символов для среза по умолчанию равно 0, а количество повторений —
1.
Функция должна возвращать полученную строку.

```python
def trim_and_repeat(input_string: str, offset: int = 0, repetitions: int = 1) -> str:
    # Шаг 1: Обрезать строку слева
    # Python-слайсинг s[start:] обрезает строку, начиная с индекса 'start'.
    # Если offset больше длины строки, результатом будет пустая строка,
    # что соответствует требуемому поведению.
    trimmed_string = input_string[offset:]

    # Шаг 2: Повторить обрезанную строку
    # Оператор * для строк в Python повторяет строку заданное количество раз.
    # Если repetitions равно 0 или отрицательно, результатом будет пустая строка.
    result_string = trimmed_string * repetitions

    return result_string
```

```python
print(f"trim_and_repeat('hello world', offset=6): '{trim_and_repeat('hello world', offset=6)}'")
```
