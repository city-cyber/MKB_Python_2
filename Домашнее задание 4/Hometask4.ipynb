Напишите функцию date_range, которая возвращает список дат за период от start_date
до end_date. Даты должны вводиться в формате YYYY-MM-DD. В случае неверного
формата или при start_date > end_date должен возвращаться пустой список.

```python
from datetime import datetime, timedelta 
```

```python
def date_range(start_date_str, end_date_str):
    date_format = "%Y-%m-%d"
    try:
        start_date = datetime.strptime(start_date_str, date_format)
        end_date = datetime.strptime(end_date_str, date_format)
    except ValueError:
        # Возвращаем пустой список, если формат даты неверный
        return []

    if start_date > end_date:
        # Возвращаем пустой список, если начальная дата позже конечной
        return []

    result = []
    current_date = start_date
    while current_date <= end_date:
        result.append(current_date.strftime(date_format))
        current_date += timedelta(days=1) # Добавляем один день
        
    return result
```

```python
начальная_дата = '2023-10-01'
конечная_дата = '2023-10-31'

результат_функции = date_range(начальная_дата, конечная_дата)

print("--- Пример вызова функции date_range ---")
print(f"Вызвана функция: date_range('{начальная_дата}', '{конечная_дата}')")
print(f"Тип возвращаемого значения: {type(результат_функции)}")
print(f"Количество дат в результате: {len(результат_функции)}")

# Выведем только часть результата, так как он может быть длинным
if len(результат_функции) > 10:
    print(f"Первые 5 дат: {результат_функции[:5]}")
    print("...")
    print(f"Последние 5 дат: {результат_функции[-5:]}")
else:
    print(f"Результат: {результат_функции}")
```
