Переведите содержимое файла purchase_log.txt в словарь purchases вида:
{'1840e0b9d4': 'Продукты', ...}
Пример работы программы при выводе первых двух элементов словаря
purchases:
1840e0b9d4 ‘Продукты‘
4e4f90fcfb ‘Электроника‘

```python
import json

purchases = {}

with open('purchase_log.txt', 'r', encoding='utf-8') as f: # читаем файл
    next(f) # пропуск первой строки заголовка
    for line in f: # перебираем строки в цикле
                purchase = json.loads(line) # построчно
                purchases[purchase['user_id']] = purchase['category'] # наполняем словарь
```

```python
for i, (user_id, category) in enumerate(purchases.items()):
    if i < int(3):
        print(f"{user_id} {category}")
    else:
        break
```

Для каждого user_id в файле visit_log.csv определите третий столбец с
категорией покупки. Если покупка была, сам файл visit_log.csv изменять не
надо.
Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были
покупки с указанием категории.
Учтите условия для данных:
● содержимое purchase_log.txt помещается в оперативную память
компьютера;
● содержимое visit_log.csv — нет, используйте только построчную
обработку этого файла.
Примеры работы программы при выводе первых трёх строк файла funnel.csv:
user_id,source,category
1840e0b9d4,other,Продукты
4e4f90fcfb,context,Электроника

```python
import json

purchases = {}

with open('purchase_log.txt', 'r', encoding='utf-8') as f: # читаем файл
    next(f) # пропуск первой строки заголовка
    for line in f: 
                purchase = json.loads(line) 
                purchases[purchase['user_id']] = purchase['category'] # наполняем словарь

with open('visit_log.csv', 'r', encoding='utf-8') as in_file:
    with open('funnel.csv', 'w', encoding='utf-8') as out_file:

        out_file.write(in_file.readline().strip() + ',category\n')

        for line in in_file:
            line = line.strip()
            user_id = line.split(',')[0]

            if purchases.get(user_id):
                category = purchases.get(user_id, '')
                out_file.write(line + ',' + category + '\n')


with open('funnel.csv', 'r', encoding='utf-8') as f:
    for i in range(4):
        line = f.readline().strip()
        print(line)
```
